```python
## get gene presence / absence
from collections import defaultdict
tcount = defaultdict(int)
pcount = defaultdict(int)
fcount = defaultdict(int)

f = open('/groups/banfield/users/alexcc/song_collaboration/metagenomic/matt_data/siderophore_gene_presence.txt')
for line in f.readlines():
    cluster = line.split()[2].replace(".gbk","")
    if int(line.split()[3]) > 0:
        pcount[cluster] += 1
    if int(line.split()[4]) > 0:
        fcount[cluster] += 1
    if int(line.split()[5]) > 0:
        tcount[cluster] += 1
f.close()


f = open('/groups/banfield/users/alexcc/song_collaboration/metagenomic/nayfach_data/siderophore_gene_presence.txt')
for line in f.readlines():
    cluster = line.split()[0] + ":" + line.split()[1]
    if int(line.split()[2]) > 0:
        pcount[cluster] += 1
    if int(line.split()[3]) > 0:
        fcount[cluster] += 1
    if int(line.split()[4]) > 0:
        tcount[cluster] += 1
f.close()

```

```python
bigscape = {}
import glob
for fn in glob.glob('/groups/banfield/users/alexcc/song_collaboration/metagenomic/bigscape_combined_final/network_files/2020-05-25_15-22-31_glocal/*/*_clustering_c0.30.tsv'):
    f = open(fn)
    print(fn)
    for line in  f.readlines():
        bgc = line.split()[0]
        if 'das' in bgc:
            bgc = bgc
        else:
            bgc = bgc.replace("-",":")
        fam = line.split()[1]
        bigscape[bgc] = fam
```

```python
f = open('/groups/banfield/users/alexcc/song_collaboration/metagenomic/combined_bigscape_matt_nayfach/cluster_types.txt')

for line in f.readlines():
    clust = line.split()[0]
    ctype = line.split()[1]
    study = 'Adult'
    if 'das' in clust:
        study = 'Infant'
        genome = line.split("-")[0]
    else:
        clust = clust.replace("-", ":")
        genome = clust.split(":")[0]
    if clust in bigscape and genome in taxa:
        big = bigscape[clust]
        tax = taxa[genome]
        print(clust + "\t" + big + "\t" + genome + "\t" + tax + "\t" + ctype + "\t" + study + "\t" + str(pcount[clust]) + "\t" + str(fcount[clust]) + "\t" + str(tcount[clust]))
f.close()
```
